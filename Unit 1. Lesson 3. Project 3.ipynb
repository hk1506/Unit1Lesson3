```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
bizdata = pd.read_csv('businesslist.csv')
```

```python
bizdata
```

```python
bizdata.groupby('Type of Business')['Zip Code'].count()
```

```python
sns.set(style='whitegrid',color_codes=True)
```

```python
sns.stripplot(y='Longitude', x='Council District',data=bizdata, jitter= True);
```

Pros:
    -  You can easily see which council district is more popular
    
Cons:
    -  It's hard to see how many council districts are on the longitude

```python
sns.violinplot(x='Zip Code', y='Longitude', data=bizdata, inner=None)
sns.swarmplot(x='Zip Code',y="Longitude", data=bizdata,color='w',alpha=.5);
```

Pro:
    -  Allows you to see the density of businesses based on location and zip code
Con: 
    -  Difficult to tell which zipcode/ area is most popular


```python
sns.factorplot(x='Zip Code', y='Longitude', data=bizdata, kind='swarm');
```

Pro: 
    -  Shows the longitude of business by zip code
Con:
    -  Not an idealistic way to use the graph

```python
sns.countplot(y="Type of Business", data=bizdata, palette="Greens_d");
```

Pros:
    -  Clean way to display data. Shows the count of businesses in the area
Con:
    -  Unable to see exactly where the business is

```python
with sns.axes_style('white'):
    sns.jointplot(x='Longitude',y='Latitude',color='k',data=bizdata)
```

Pro:
    -  Able to see the grid like structure of NYC

Con: 
    -  The dots a little too large and has a lot of overlap

```python
sns.jointplot(x='Longitude',y='Latitude',data=bizdata,kind='kde');
```

Pro: 
    -  Shows which location is the most populated
Con:
    -  Density is relative to the total-- therefore, you can't see exactly how many businesses are present

```python
g = sns.jointplot(x="Longitude", y="Latitude", data=bizdata, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

Pro: 
    -  Shows the popular locations by density and by exact location 
    -  Shows the grid like map of NYC 
    
Con:
    -  It's very crowded and can get difficult to read

```python
sns.stripplot(x='Longitude', y='Type of Business',data=bizdata, jitter= True)
```

Pro: 
    -  Able to see which kind of businesses are the most popular 
    -  Colors help distinguish different businesses
Con: 
    -  Difficult to see how numerical differences between businesses
    -  Not an idealistic use of plot 

```python
sns.factorplot(x="Zip Code",y="Longitude", data=bizdata);
```

Pro: 
    -  Provides a general idea of the start and end longitude of the Union Sqaure Area
Con:
    -  Does not display all the longitudes

```python
sns.pointplot(x="Zip Code", y="Longitude", hue="Zip Code", data=bizdata);
```

Pro:
    -  Easy to read and clean
Con:
    -  Doesn't provide all all the longitudes for the zip code
